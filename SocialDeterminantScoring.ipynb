Assign Quantiles to Each County for each variable in Health


```python
import pandas as pd
from collections import Counter

# Assuming Health is your DataFrame
# Loop through each numerical column in the DataFrame
for column in Health.select_dtypes(include=['number']).columns:
    if column != 'HSPA':  # Skip the 'HSPA' column to avoid the error
        quantile_column = f'{column}_quantile'
        # Calculate the quantiles without assigning to DataFrame first to get the number of bins
        try:
            quantiles, bins = pd.qcut(Health[column], q=4, retbins=True, duplicates='drop')
            labels = [f'Q{i+1}' for i in range(len(bins)-1)]  # Create labels based on number of bins
            Health[quantile_column] = pd.qcut(Health[column], q=4, labels=labels, duplicates='drop')
        except ValueError as e:
            print(f"Skipping column {column} due to error: {e}")

# Dictionary to store counties for specific quantiles of specific columns
counties_with_Q1_Q4 = {
    'Life Expectancy_quantile': [],
    'Primary Care Physicians Ratio Numeric_quantile': [],
    'Pre-Mature Mortality Rate (death before age 75 per 100k)_quantile': [],
    '% Uninsured Children_quantile': [],
    '% Uninsured Adults_quantile': []
}

# Store counties with 'Q1' for specified columns and 'Q4' for others
for quantile_col in counties_with_Q1_Q4.keys():
    if quantile_col in Health.columns:
        if 'Life Expectancy' in quantile_col or 'Primary Care Physicians Ratio Numeric' in quantile_col:
            counties_with_Q1_Q4[quantile_col] = Health[Health[quantile_col] == 'Q1']['COUNTY'].tolist()
        elif 'Pre-Mature Mortality Rate (death before age 75 per 100k)' in quantile_col or '% Uninsured Children' in quantile_col or '% Uninsured Adults' in quantile_col:
            counties_with_Q1_Q4[quantile_col] = Health[Health[quantile_col] == 'Q4']['COUNTY'].tolist()

# Print the results
for quantile_col, counties in counties_with_Q1_Q4.items():
    quantile = 'Q1' if 'Life Expectancy' in quantile_col or 'Primary Care Physicians Ratio Numeric' in quantile_col else 'Q4'
    print(f'Counties with {quantile} in {quantile_col}: {counties}')

# Flatten the lists of counties and count occurrences
all_counties = [county for counties in counties_with_Q1_Q4.values() for county in counties]
county_counts = Counter(all_counties)

# Print counties that appear in 2 or more quantile lists
counties_in_multiple_quantiles = [county for county, count in county_counts.items() if count >= 2]

print("Counties in 2 or more quantile lists:", counties_in_multiple_quantiles)

Poor_Health_counties = Health[Health['COUNTY'].isin(counties_in_multiple_quantiles)]

print(Poor_Health_counties)

```

Health Scoring Grid

---

Weightings:

*   Life Expectancy  -- 1
*   Pre- Mature Mortality Rate (death before age 75 per 100k) -- 1
*   % Uninsured Adults -- 2
*   % Uninsured Children -- 1
*   Primary Care Physician Ratio Numeric -- 2

```python
# Function to map quantile to score
def quantile_to_score(quantile):
    quantile_map = {'Q4': 4, 'Q3': 3, 'Q2': 2, 'Q1': 1}
    return quantile_map.get(quantile, 0)  # Return 0 if quantile not found

# Calculate Life Expectancy Score
Health['Life Expectancy_Score'] = Health['Life Expectancy_quantile'].map(quantile_to_score)



# Function to map quantile to score
def quantile_to_score2(quantile):
    quantile_map = {'Q4': 1, 'Q3': 2, 'Q2': 3, 'Q1': 4}
    return quantile_map.get(quantile, 0)  # Return 0 if quantile not found

# Calculate Primary Care Physician Ratio Score
Health['Primary Care Physician Ratio Numeric_Score'] = Health['Primary Care Physicians Ratio Numeric_quantile'].map(quantile_to_score2)

# Calculate % Uninsured Adult Score
Health['% Uninsured Adults_Score'] = Health['% Uninsured Adults_quantile'].map(quantile_to_score2).astype(float)

# Calculate % Uninsured Children Score
Health['% Uninsured Children_Score'] = Health['% Uninsured Children_quantile'].map(quantile_to_score2).astype(float)

#Calculate Pre-Mature Mortality rate score

Health['Pre-Mature Mortality Rate (death before age 75 per 100k)_Score'] = Health['Pre-Mature Mortality Rate (death before age 75 per 100k)_quantile'].map(quantile_to_score2).astype(float)

# Define weights
weights = {
    '% Uninsured Children_Score': 1,
    '% Uninsured Adults_Score': 2,
    'Primary Care Physician Ratio Numeric_Score': 2,
    'Life Expectancy_Score': 1,
    'Pre-Mature Mortality Rate (death before age 75 per 100k)_Score': 1
}

# Normalize weights so their sum equals 1 (optional)
total_weight = sum(weights.values())
weights = {key: val / total_weight for key, val in weights.items()}

# Convert score columns to numeric type before calculation
for col in weights.keys():
    Health[col] = pd.to_numeric(Health[col], errors='coerce')  # Convert to numeric, handle errors if any

# Calculate Composite Score using the specified weights
Health['Composite_Score'] = sum(Health[col] * weight for col, weight in weights.items())
Health['Data Category'] = 'Health'
display(Health)

Health.to_excel('Health_Data.xlsx', index=False)
```

```python

# Function to combine # and % columns
def combine_columns(df):
    new_columns = {}
    for col in df.columns:
        if col.startswith('#'):
            percentage_col = col.replace('#', '%').strip()
            if percentage_col in df.columns:
                new_column_name = col.replace('#', '').strip()
                new_columns[new_column_name] = df[col].astype(str) + " (" + df[percentage_col].astype(str) + "%)"

    # Drop the old columns and add the new combined columns
    df = df.drop(columns=[col for col in df.columns if col.startswith('#') or col.startswith('%')])
    for col_name, new_col_data in new_columns.items():
        df[col_name] = new_col_data

    return df

# Modify Demographics_modified DataFrame
Demographics_modified = combine_columns(Demographics)
display(Demographics_modified)
# Drop Demographics_modified columns
Demographics_modified= Demographics_modified.drop(columns=['HPSA','Region','Primary Care Physicians Ratio Numeric'])
display(Demographics_modified)

display(Demographics)
```

Nutrition Scoring

```python
# Function to map quantile to score
def quantile_to_score(quantile):
    quantile_map = {'Q4': 1, 'Q3': 2, 'Q2': 3, 'Q1': 4}
    return quantile_map.get(quantile, 0)  # Return 0 if quantile not found

# Apply quantile_to_score to columns 0 and 3
target_columns_score = ['% Food Insecure_quantile', '% Adults with Obesity_quantile',
       '% Limited Access to Healthy Foods_quantile']
for column_name in target_columns_score: # Iterate directly over column names
    Nutrition[f'{column_name}_Score'] = Nutrition[column_name].map(quantile_to_score)

# Recalculate scores for Nutrition dataset
# Define weights for each variable in the Composite Score
weights = {
     '% Food Insecure_quantile_Score': 2,
       '% Adults with Obesity_quantile_Score':1,
       '% Limited Access to Healthy Foods_quantile_Score': 1
}

# Convert score columns to numeric type BEFORE calculation
for col in weights.keys():
    if col in Nutrition.columns:  # Check if the column exists before conversion
        Nutrition[col] = pd.to_numeric(Nutrition[col], errors='coerce')

# Normalize weights so their sum equals 1 (optional, but recommended for comparability)
total_weight = sum(weights.values())
weights = {key: val / total_weight for key, val in weights.items()}

# Calculate the Composite Score using the specified weights row by row
Nutrition['Composite_Score'] = Nutrition.apply(
    lambda row: round(sum(row[col] * weight for col, weight in weights.items() if pd.notnull(row[col])), 2), axis=1
)


# Save the result to Excel
Nutrition.to_excel('Nutrition_Data.xlsx', index=False)

# Display the DataFrame
display(Nutrition)
```
